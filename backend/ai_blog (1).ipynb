```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Load model and tokenizer from Hugging Face
model_name = "gpt2-large"  # You can also try 'EleutherAI/gpt-neo-1.3B' if you have more RAM

tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
model.eval()  # Inference mode

# Function to generate a blog
def generate_blog(prompt, max_length=300, temperature=0.8, top_p=0.95):
    # Tokenize input prompt
    input_ids = tokenizer.encode(prompt, return_tensors="pt")

    # Generate output
    with torch.no_grad():
        output = model.generate(
            input_ids,
            do_sample=True,            # enables sampling
            max_length=max_length,     # max length of generated text
            top_p=top_p,               # nucleus sampling
            temperature=temperature,   # randomness
            pad_token_id=tokenizer.eos_token_id,
        )

    # Decode generated tokens
    blog = tokenizer.decode(output[0], skip_special_tokens=True)
    return blog

# Example prompt
prompt = "Write a blog post about how AI is transforming the retail industry:\n\n"

# Generate blog
blog_post = generate_blog(prompt)
print(blog_post)
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel, Trainer, TrainingArguments
from datasets import Dataset
import json
import os
os.environ["WANDB_DISABLED"] = "true"
# Load GPT-2 and tokenizer
model = GPT2LMHeadModel.from_pretrained("gpt2")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
tokenizer.pad_token = tokenizer.eos_token  # Set pad token

# Load dataset
texts = []
titles = []

with open('/content/data_combined.json', 'r') as f:
    for line in f:
        item = json.loads(line.strip())
        titles.append(item['title'])
        texts.append(item['text'])

# Combine title and text
combined_texts = [f"Title: {title}\nText: {text}" for title, text in zip(titles, texts)]

# Create Hugging Face dataset
dataset = Dataset.from_dict({"text": combined_texts})

# Tokenization function
def tokenize_function(examples):
    outputs = tokenizer(
        examples["text"],
        truncation=True,
        padding="max_length",
        max_length=512,
    )
    outputs["labels"] = outputs["input_ids"].copy()  # For causal LM
    return outputs

# Tokenize the dataset
dataset = dataset.map(tokenize_function, batched=True, remove_columns=["text"])

# Training arguments
training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=3,
    per_device_train_batch_size=2,
    save_steps=10_000,
    save_total_limit=2,
    logging_dir="./logs",
    logging_steps=500,
    prediction_loss_only=True,
)

# Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset,
)

# Train
trainer.train()

# Save model and tokenizer
model.save_pretrained("./fine_tuned_model")
tokenizer.save_pretrained("./fine_tuned_model")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# Save the fine-tuned model and tokenizer
model.save_pretrained("./mistral_model")  # Save the model
tokenizer.save_pretrained("./mistral_model")  # Save the tokenizer

# Load the fine-tuned model and tokenizer
model = AutoModelForCausalLM.from_pretrained("./mistral_model")
tokenizer = AutoTokenizer.from_pretrained("./mistral_model")

# Set the padding token to be EOS (important if it was set earlier)
tokenizer.pad_token = tokenizer.eos_token

# Set model to evaluation mode
model.eval()

# Input prompt (title or title+text)
prompt = "Title: AI in Healthcare\nText:"

# Tokenize the input prompt
input_ids = tokenizer(prompt, return_tensors="pt").input_ids

# Generate continuation
generated_ids = model.generate(
    input_ids,
    max_length=1000,  # Maximum length of the generated text
    num_return_sequences=1,  # Number of sequences to return
    pad_token_id=tokenizer.eos_token_id,  # Padding token (eos_token is used)
    do_sample=True,  # Enables sampling (instead of greedy search)
    top_p=0.95,  # Nucleus sampling (probability mass)
    temperature=0.7  # Sampling temperature (controls randomness)
)

# Decode and print the generated text
generated_text = tokenizer.decode(generated_ids[0], skip_special_tokens=True)
print(generated_text)
```
